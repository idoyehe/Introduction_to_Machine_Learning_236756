## <a name="k_means"></a>K-Means Clustering

We already used the iris dataset. Let us read it. 

```python
iris = pd.read_csv('../Data/iris.csv')
iris.head(10)
```

```python
iris["Species"].value_counts()
```

Let us pairplot the numeric features of the dataset, and denote the different species with different colors.

```python
sns.pairplot(iris,hue="Species");
```

It appears that the different species are destributed differently in the input space. Let us use K-Means clustering to partition the space.

```python
random_state = 170
clusters = KMeans(n_clusters=3, random_state=random_state).fit_predict(iris.drop(['Species'],axis = 1))
iris['clusters'] = clusters
```

Let us compute the centers of each cluster. And examine the seperation of the input space.

```python
iris.groupby("clusters").mean()[["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]]
```

Let's plot again, but this time let's color it based on the cluster it belongs to

```python
#fig, ax = plt.subplots()
# the size of A4 paper
#fig.set_size_inches(11.7, 8.27)
sns.pairplot(iris,hue="clusters")
```

Can we use the clusters for explaining the species?

```python
conf_matrix = pd.crosstab(iris.Species,iris.clusters)
print(conf_matrix)
print(np.array(conf_matrix).trace()/np.array(conf_matrix).sum())
```

In principle, one can encode several input features as clusters and use the cluster relation as an explaining feature.

## <a name="gmm"></a>GMMs

```python
iris = pd.read_csv('../Data/iris.csv')
iris_explain = iris.drop(['Species'],axis = 1)
iris_target = iris.Species 
```

```python
colors = ['crimson', 'green', 'darkorange']
classes = ["setosa","versicolor","virginica"] 
n_classes = len(np.unique(iris_target))
#cov_type = 'spherical'
cov_type = 'full'
estimator = GaussianMixture(n_components=n_classes,covariance_type=cov_type, max_iter=20, random_state=0)
estimator.means_init = np.array([iris_explain[iris_target == i].mean(axis=0) for i in ["setosa","versicolor","virginica"]])
estimator.fit(iris_explain)
plt.figure(figsize=(10,5))
plt.subplots_adjust(bottom=.01, top=0.95, hspace=.15, wspace=.05,left=.01, right=.99)
h = plt.subplot(1, 1, 1)
for n, color in enumerate(colors):
    #covariances = np.eye(estimator.means_.shape[1]) * estimator.covariances_[n]
    covariances = estimator.covariances_[n][:2, :2]
    v, w = np.linalg.eigh(covariances)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan2(u[1], u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    v = 2. * np.sqrt(2.) * np.sqrt(v)
    ell = mpl.patches.Ellipse(estimator.means_[n, :2], v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(h.bbox)
    ell.set_alpha(0.5)
    h.add_artist(ell)
for n, color in enumerate(colors):
    data = iris[iris.Species == classes[n]]
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], s=4, color=color,label=iris.Species[n])
    # Plot the test data with crosses


plt.title("GMM on Iris")
plt.xlabel(data.columns[0])
plt.ylabel(data.columns[1])
plt.show()
```

## <a name="dbscan"></a>K-Means vs. DBSCAN

K-Means is a very simple and straight forward algorithm. It have few limitations. One of them is the way it seperates the space. K-Means produces hyper polygons in the input space. This way of partitioning ignores the real distribution of the data.

Let us produce a synthetic dataset.

```python
X,y = make_circles(n_samples=400, factor=.3, noise=.05)
circle = pd.DataFrame(X)
sns.pairplot(circle)
```

Clearly, the type of partition that we would expect is the inner loop vs. the outer loop. Unless we transform the input space, K-means cannot achieve such partition.

```python
clusters = KMeans(n_clusters=2, random_state=random_state).fit_predict(circle)
circle["clusters"] = clusters
sns.pairplot(circle,hue="clusters")
```

DBSCAN (Density-based spatial clustering of applications with noise) is a data clustering algorithm that focus on the real distribution of the data. It groups together data points that are close in the input space (it basically start with a point and strive to cluster it with points that are close to it).

```python
random_state = 170
DB_clusters = DBSCAN(eps=0.5,min_samples=50).fit_predict(circle)
circle["clusters"] = DB_clusters
sns.pairplot(circle,hue="clusters")
```

## <a name="arules"></a>Association Rules

```python
retail_transactions = pd.read_excel('../Data/Online_Retail.xlsx')
retail_transactions.head()
```

```python
retail_transactions.shape
```

Let us run some pre-processing on the data

```python
# Remove white spaces ebfor and after the item description string
retail_transactions['Description'] = retail_transactions['Description'].str.strip()
# Convers InvoiceNo to string
retail_transactions['InvoiceNo'] = retail_transactions['InvoiceNo'].astype('str')
```

```python
retail_transactions.shape
```

Let us focus on transaction from France

```python
basket = (retail_transactions[retail_transactions['Country'] =="France"].groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
print(basket.head())
print(basket.shape)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```

## <a name="word2vec"></a>Word2Vec
The idea behind Word2Vec is pretty simple. Were making an assumption that the meaning of a word can be inferred by the company it keeps.  
The secret to getting Word2Vec really working for you is to have lots and lots of text data in the relevant domain.  
For example, if your goal is to build a sentiment lexicon, then using a dataset from the medical domain or even wikipedia may not be effective. So, choose your dataset wisely.  
We are going to use a dataset which user reviews of cars and hotels (taking from http://kavita-ganesan.com/entity-ranking-data/#.WxNXclOFMmU).  

For this practice we will use **Gensim** package.  
Gensim is a robust open-source vector space modeling and topic modeling toolkit implemented in Python. It uses NumPy, SciPy and optionally Cython for performance. Gensim is specifically designed to handle large text collections, using data streaming and efficient incremental algorithms, which differentiates it from most other scientific software packages that only target batch and in-memory processing.

```python
# imports needed and logging
import gzip
import gensim # might need to install first
import logging
 
input_file = "../Data/reviews_data.txt.gz"
# loading data
with gzip.open (input_file, 'rb') as f:
        for i,line in enumerate (f):
            print(line)
            break;

def read_input(input_file):
    """This method reads the input file which is in gzip format"""
 
    logging.info("reading file {0}...this may take a while".format(input_file))
    with gzip.open(input_file, 'rb') as f:
        for i, line in enumerate(f):
 
            if (i % 10000 == 0):
                logging.info("read {0} reviews".format(i))
            # do some pre-processing and return list of words for each review
            # text
            yield gensim.utils.simple_preprocess(line) # Convert a document into a list of tokens. This lowercases, tokenizes, de-accents (optional).
            
```

Now let's read it all this time

```python
data_file = "../Data/reviews_data.txt.gz"

# read the tokenized reviews into a list
# each review item becomes a serries of words
# so this becomes a list of lists
documents = list(read_input(data_file))
logging.info("Done reading data file")
```

```python
len(documents)
```

Train the network

```python
model = gensim.models.Word2Vec(
    documents,
    size=150,
    window=10,
    min_count=2,
    workers=10)
model.train(documents, total_examples=len(documents), epochs=10)
```

And now the fun begin :)
![alt_text](https://www.tensorflow.org/images/linear-relationships.png)

```python
# save only the word vectors
#model.wv.save(os.path.join(abspath, "../vectors/default"))

w1 = "dirty"
print("Most similar to {0}".format(w1), model.wv.most_similar(positive=w1))

# look up top 6 words similar to 'polite'
w1 = ["polite"]
print(
    "Most similar to {0}".format(w1),
    model.wv.most_similar(
        positive=w1,
        topn=6))

# look up top 6 words similar to 'france'
w1 = ["france"]
print(
    "Most similar to {0}".format(w1),
    model.wv.most_similar(
        positive=w1,
        topn=6))

# look up top 6 words similar to 'shocked'
w1 = ["shocked"]
print(
    "Most similar to {0}".format(w1),
    model.wv.most_similar(
        positive=w1,
        topn=6))

# look up top 6 words similar to 'shocked'
w1 = ["beautiful"]
print(
    "Most similar to {0}".format(w1),
    model.wv.most_similar(
        positive=w1,
        topn=6))

# get everything related to stuff on the bed
w1 = ["bed", 'sheet', 'pillow']
w2 = ['couch']
print(
    "Most similar to {0}".format(w1),
    model.wv.most_similar(
        positive=w1,
        negative=w2,
        topn=10))

# similarity between two different words
print("Similarity between 'dirty' and 'smelly'",
      model.wv.similarity(w1="dirty", w2="smelly"))

# similarity between two identical words
print("Similarity between 'dirty' and 'dirty'",
      model.wv.similarity(w1="dirty", w2="dirty"))

# similarity between two unrelated words
print("Similarity between 'dirty' and 'france'",
      model.wv.similarity(w1="dirty", w2="france"))
```
