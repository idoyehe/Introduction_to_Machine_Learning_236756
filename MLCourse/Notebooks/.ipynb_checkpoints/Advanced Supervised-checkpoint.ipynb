## <a name="dataset_description"></a>Titanic Dataset Description:

The **Titanic** data set contain different information about passengers onboard the Titanic, and we see if we can use that information to predict whether those people survived or not.

Features:
* survival - Survival (0 = No; 1 = Yes)
* class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
* name - Name
* sex - Sex
* age - Age
* sibsp - Number of Siblings/Spouses Aboard
* parch - Number of Parents/Children Aboard
* ticket - Ticket Number
* fare - Passenger Fare
* cabin - Cabin
* embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time

from sklearn.preprocessing import scale
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from IPython.display import display
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import roc_curve, auc
from sklearn.svm import SVC

# Importing dataset
orig_data = pd.read_csv("../Data/titanic.csv")
orig_data.head()
```

preparing a ROC draw function

```python
orig_data.shape
```

```python
def drawROC(X_test, *args):  
    # A place holder for the ROC plot
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic for Survived')
    
    idx = 0
        
    for model in args:
 
        model_name = model[0]
        model_probs = model[1]
        
        probs = model_probs[:,1]
        fpr, tpr, _ = roc_curve(X_test["Survived"]==1, probs) # survived == true
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, color='C'+str(idx),
             lw=2, label='{} ROC curve (area = {})'.format(model_name, round(roc_auc,2)))
        idx = idx + 1
                 
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='b',label='Luck', alpha=.8)
    plt.legend(loc="lower right")
```

## <a name="naive_bayes"></a>Naive Bayes
The Naive Bayes classifier aggregates information using conditional probability with an assumption of independence among features.

Let's start with some basic pre-processing

```python
data = orig_data.copy()

# Convert categorical variable to numeric
data["Sex_cleaned"]=np.where(data["Sex"]=="male",0,1)
data["Embarked_cleaned"]=np.where(data["Embarked"]=="S",0,
                                  np.where(data["Embarked"]=="C",1,
                                           np.where(data["Embarked"]=="Q",2,3)
                                          )
                                 )
# Cleaning dataset of NaN
data_cleaned = data[[
    "Survived",
    "Pclass",
    "Sex_cleaned",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked_cleaned"
]].dropna(axis=0, how='any').copy()

print(data_cleaned.shape)
data_cleaned.head()
```

Let's start with one explaining variable

```python
# Define the model - you need to make some assumption regarding the distribution
nb = GaussianNB()

# Split dataset in training and test datasets
X_train, X_test = train_test_split(data_cleaned, test_size=0.3, random_state=42)
used_features = ["Pclass"]

# Train classifier
nb.fit(X_train[used_features].values,X_train["Survived"])

y_pred = nb.predict(X_test[used_features])
print(y_pred[:5])

# Print results
c_matrix = confusion_matrix(X_test["Survived"], y_pred)
print(nb.classes_)
print(c_matrix)
dt_acc = c_matrix.trace()/c_matrix.sum()
print("Total accuracy: {}".format(dt_acc))
print("Survived Recall: {}".format(c_matrix[1,1]/sum(c_matrix[1,:])))
print("Survived Precision: {}".format(c_matrix[1,1]/sum(c_matrix[:,1])))
```

Accuracy is bad. Now let's use more features

```python
nb = GaussianNB()
used_features =[
    "Pclass",
    "Sex_cleaned",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked_cleaned"
]

# Train classifier
nb.fit(
    X_train[used_features].values,
    X_train["Survived"]
)
```

And now predict

```python
y_pred = nb.predict(X_test[used_features])
print(y_pred[:5])

# Print results
c_matrix = confusion_matrix(X_test["Survived"], y_pred)
print(nb.classes_)
print(c_matrix)
dt_acc = c_matrix.trace()/c_matrix.sum()
print("Total accuracy: {}".format(dt_acc))
```

Let's draw an ROC

```python
nb_pred_prob = nb.predict_proba(X_test[used_features])
print(nb_pred_prob[0:10,:])
drawROC(X_test, ('Naive Bayes', nb_pred_prob));
```

## <a name="logistic_regression"></a>Logistic Regression

Let's start with a default LR model

```python
X_train, X_test = train_test_split(data_cleaned, test_size=0.3, random_state=42)

lr = LogisticRegression(random_state=1)
lr = lr.fit(X_train[used_features], X_train["Survived"])
print(lr.coef_)

# predict
lr_prob_predict = lr.predict_proba(X_test[used_features])

drawROC(X_test, ('Default LR', lr_prob_predict))
```

Let's see how our default model was defined

```python
print(lr)
```

Notice that we have a regularization parameter (C). The smaller C is the stronger is the effect of regularization.  
The two most common methods are l1 (lasso) and l2 (ridge). The default is l2.


Let's play with the regularization parameters. 

```python
lr = LogisticRegression(penalty='l2', C=0.05, random_state=1)
lr = lr.fit(X_train[used_features], X_train["Survived"])
print(lr.coef_)

# predict
lr_adujsted_prob_predict = lr.predict_proba(X_test[used_features])
drawROC(X_test, ('Adjusted LR', lr_adujsted_prob_predict), ('Default LR', lr_prob_predict))
print(used_features)
```

```python
lr = LogisticRegression(penalty='l1', C=0.05, random_state=1)
lr = lr.fit(X_train[used_features], X_train["Survived"])
print(lr.coef_)

# predict
lr_adujsted_prob_predict = lr.predict_proba(X_test[used_features])
drawROC(X_test, ('Adjusted LR', lr_adujsted_prob_predict), ('Default LR', lr_prob_predict))
print(used_features)
```

The explaining variables that survived the lasso are: Pclass, Sex_cleaned, Age anf Fare

```python
lr = LogisticRegression(penalty='l2', C=0.05, random_state=1)
lr = lr.fit(X_train[used_features], X_train["Survived"])
print(lr.coef_)

# predict
lr_ridge_adujsted_prob_predict = lr.predict_proba(X_test[used_features])
drawROC(X_test, ('Adjusted LR', lr_ridge_adujsted_prob_predict), ('Default LR', lr_prob_predict))
print(used_features)
```

Ridge is striving to makethe coeficients small (not necessarily 0). Lasso is more probable to select variables.

## <a name="svm"></a>SVM

Let's start with the default kernel which is RBF

```python
# rbf svm
svc = SVC(probability=True) # default SVM, need flag for probability
svc
```

The default is a radial basis function

```python
svc.fit(X_train[used_features], X_train["Survived"]) 
svc_rbf_prob_predict = svc.predict_proba(X_test[used_features])

drawROC(X_test, ('RBF SVM',svc_rbf_prob_predict))
```

Now let's try a linear SVM

```python
# linear svm
svc = SVC(kernel='linear', probability=True) # need flag for probability
svc.fit(X_train[used_features], X_train["Survived"]) 
svc_linear_prob_predict = svc.predict_proba(X_test[used_features])

drawROC(X_test, ('RBF SVM',svc_rbf_prob_predict), ('Linear SVM',svc_linear_prob_predict))
```

## <a name="random_forest"></a>Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth = 5, min_samples_leaf=10,n_estimators=10 ,random_state=0)
rf.fit(X_train[used_features], X_train["Survived"])
print(rf)
```

Let's predict

```python
# predict
rf_prob_predict = rf.predict_proba(X_test[used_features])

drawROC(X_test, ('Random Forest',rf_prob_predict))
```

Feature importance using random forest

```python
importance = rf.feature_importances_
importance = pd.DataFrame(importance, index=X_train[used_features].columns, 
                          columns=["Importance"])

importance["Std"] = np.std([rf.feature_importances_
                            for tree in rf.estimators_], axis=0)


sort_imp = importance.sort_values(by=['Importance'], ascending=False)

print(sort_imp)
```

The top important features are those who survived the lasso

## <a name="nueral_networks"></a>Neural Networks
scikit has it's own NN implementation, howerver when it come to NN, the most used packages are:
* tensorflow
* pytorch
* keras (HL api for tensorflow)

We will use keras.

```python
from __future__ import print_function
import numpy as np
import pandas as pd
from keras.models import Sequential # should install keras + tensorflow first
from keras.optimizers import SGD, RMSprop, Adam
from keras.layers import Dense, Activation, Dropout
```

Let's start with a simple 2 hidden layer net

```python
# create model
model = Sequential() # A feed-forward multi-layer network
model.add(Dense(input_dim=X_train[used_features].shape[1], units=16,
                 kernel_initializer='normal', bias_initializer='zeros')) # defining the first layer
model.add(Activation('relu'))

k=1
for i in range(0, k): # adding K more layers
    model.add(Dense(units=16, kernel_initializer='normal',
                     bias_initializer='zeros'))
    model.add(Activation('relu'))

model.add(Dense(units=2)) # Adding the output
model.add(Activation('softmax'))
```

Compiling will create the network structure. After compiling we can investigate the network

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

Similar to scikit, keras will have it's own fit \ predict api

```python
y_train=pd.get_dummies(X_train["Survived"])
y_test = pd.get_dummies(X_test["Survived"])
model.fit(X_train[used_features].values, y_train.values , epochs=50, verbose=2)
```

We can get indication about the performance of our model as it learn on the test dataset

```python
y_train=pd.get_dummies(X_train["Survived"])
y_test = pd.get_dummies(X_test["Survived"])
model.fit(X_train[used_features].values, y_train.values , epochs=50, verbose=2, 
          validation_data=(X_test[used_features].values, y_test.values))
```

```python
p_nn_survived = model.predict(X_test[used_features].values)
#p_survived
drawROC(X_test, ('NN',p_nn_survived))
```

Let's draw ROC curve

```python
drawROC(X_test, ('Random Forest',rf_prob_predict), ('NN',p_nn_survived), ('Adjusted LR', lr_adujsted_prob_predict), 
        ('Default LR', lr_prob_predict), ('RBF SVM',svc_rbf_prob_predict), ('Linear SVM',svc_linear_prob_predict),
       ('Naive Bayes', nb_pred_prob))
```
